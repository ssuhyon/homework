```python
# Person í´ë˜ìŠ¤ ì •ì˜
class Person:
    def __init__(self, name, gender, age):  # ìƒì„±ì: Person ê°ì²´ë¥¼ ì´ˆê¸°í™”í•˜ëŠ” ë©”ì„œë“œ
        self.name = name  # ì…ë ¥ë°›ì€ ì´ë¦„ì„ ê°ì²´ì˜ ì†ì„±ìœ¼ë¡œ ì €ì¥
        self.gender = gender.capitalize()  # ì„±ë³„ì„ ì²« ê¸€ìë§Œ ëŒ€ë¬¸ìë¡œ ë³€í™˜í•˜ì—¬ ì €ì¥ ('male' -> 'Male')
        self.age = age  # ì…ë ¥ë°›ì€ ë‚˜ì´ë¥¼ ê°ì²´ì˜ ì†ì„±ìœ¼ë¡œ ì €ì¥

    def display(self):  # ê°ì²´ ì •ë³´ë¥¼ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ
        print(f"ì´ë¦„: {self.name}, ì„±ë³„: {self.gender}")  # ì´ë¦„ê³¼ ì„±ë³„ì„ ì¶œë ¥
        print(f"ë‚˜ì´: {self.age}")  # ë‚˜ì´ë¥¼ ì¶œë ¥

    def greet(self):  # ë‚˜ì´ì— ë”°ë¼ ë‹¤ë¥¸ ì¸ì‚¬ë§ì„ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ
        if self.age < 13:  # ë‚˜ì´ê°€ 13ì„¸ ë¯¸ë§Œì¸ ê²½ìš°
            print(f"ì•ˆë…•í•˜ì„¸ìš”, {self.name}ì–´ë¦°ì´! ë°˜ê°€ì›Œìš”!")  # ì–´ë¦°ì´ì— ëŒ€í•œ ì¸ì‚¬ë§ ì¶œë ¥
        elif 13 <= self.age <= 19:  # ë‚˜ì´ê°€ 13~19ì„¸ì¸ ê²½ìš°
            print(f"ì•ˆë…•í•˜ì„¸ìš”, {self.name}ë‹˜! ì²­ì†Œë…„ì´ì‹œêµ°ìš”!")  # ì²­ì†Œë…„ì— ëŒ€í•œ ì¸ì‚¬ë§ ì¶œë ¥
        elif 20 <= self.age <= 59:  # ë‚˜ì´ê°€ 20~59ì„¸ì¸ ê²½ìš°
            print(f"ì•ˆë…•í•˜ì„¸ìš”, {self.name}ë‹˜! ì„±ì¸ì´ì‹œêµ°ìš”!")  # ì„±ì¸ì— ëŒ€í•œ ì¸ì‚¬ë§ ì¶œë ¥
        else:  # ë‚˜ì´ê°€ 60ì„¸ ì´ìƒì¸ ê²½ìš°
            print(f"ì•ˆë…•í•˜ì„¸ìš”, {self.name}ë‹˜! ì–´ë¥´ì‹ ì´ì‹œêµ°ìš”!")  # ì–´ë¥´ì‹ ì— ëŒ€í•œ ì¸ì‚¬ë§ ì¶œë ¥


# ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
name = get_name()  # ì´ë¦„ ì…ë ¥ë°›ê¸° í•¨ìˆ˜ í˜¸ì¶œ (ìœ íš¨ì„± ê²€ì‚¬ëŠ” get_name í•¨ìˆ˜ì—ì„œ ì²˜ë¦¬)
gender = get_gender()  # ì„±ë³„ ì…ë ¥ë°›ê¸° í•¨ìˆ˜ í˜¸ì¶œ (ìœ íš¨ì„± ê²€ì‚¬ëŠ” get_gender í•¨ìˆ˜ì—ì„œ ì²˜ë¦¬)
age = get_age()  # ë‚˜ì´ ì…ë ¥ë°›ê¸° í•¨ìˆ˜ í˜¸ì¶œ (ìœ íš¨ì„± ê²€ì‚¬ëŠ” get_age í•¨ìˆ˜ì—ì„œ ì²˜ë¦¬)

# Person ê°ì²´ ìƒì„± ë° ë™ì‘
person = Person(name, gender, age)  # ì…ë ¥ë°›ì€ ë°ì´í„°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ Person ê°ì²´ ìƒì„±
person.display()  # ê°ì²´ ì •ë³´ ì¶œë ¥ ë©”ì„œë“œ í˜¸ì¶œ
person.greet()  # ë‚˜ì´ì— ë”°ë¼ ë‹¤ë¥¸ ì¸ì‚¬ë§ì„ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ í˜¸ì¶œ
```

```python
import random  # ëœë¤ ìˆ«ì ìƒì„±ì„ ìœ„í•œ random ëª¨ë“ˆ ê°€ì ¸ì˜¤ê¸°

# ìˆ«ì ë§ì¶”ê¸° ê²Œì„ í•¨ìˆ˜ ì •ì˜
def number_game():
    # ê²Œì„ ì‹œì‘ ì•ˆë‚´ ë©”ì‹œì§€ ì¶œë ¥
    print("ìˆ«ì ë§ì¶”ê¸° ê²Œì„ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!\n1ë¶€í„° 10 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”.")

    while True:  # ì‚¬ìš©ìê°€ ê²Œì„ì„ ëë‚¼ ë•Œê¹Œì§€ ë°˜ë³µ
        # 1ë¶€í„° 10 ì‚¬ì´ì˜ ëœë¤ ìˆ«ì ìƒì„±
        target_number = random.randint(1, 10)  # ì •ë‹µ ìˆ«ì ìƒì„±
        guessed_correctly = False  # ì •ë‹µì„ ë§ì·„ëŠ”ì§€ ì—¬ë¶€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” í”Œë˜ê·¸ ë³€ìˆ˜

        while not guessed_correctly:  # ì •ë‹µì„ ë§ì¶œ ë•Œê¹Œì§€ ë°˜ë³µ
            try:
                # ì‚¬ìš©ìë¡œë¶€í„° ìˆ«ìë¥¼ ì…ë ¥ë°›ìŒ
                user_input = int(input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))  # ì…ë ¥ê°’ì„ ì •ìˆ˜ë¡œ ë³€í™˜
                # ì…ë ¥ê°’ì´ 1~10 ë²”ìœ„ë¥¼ ë²—ì–´ë‚œ ê²½ìš°
                if user_input < 1 or user_input > 10:
                    print("âš ï¸ 1ë¶€í„° 10 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”.")  # ë²”ìœ„ ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥
                # ì…ë ¥ê°’ì´ ì •ë‹µë³´ë‹¤ ì‘ì€ ê²½ìš°
                elif user_input < target_number:
                    print("ì´ê²ƒë³´ë‹¨ ì»¤ìš”!")  # íŒíŠ¸ ì œê³µ
                # ì…ë ¥ê°’ì´ ì •ë‹µë³´ë‹¤ í° ê²½ìš°
                elif user_input > target_number:
                    print("ì´ê²ƒë³´ë‹¨ ì‘ì•„ìš”!")  # íŒíŠ¸ ì œê³µ
                # ì…ë ¥ê°’ì´ ì •ë‹µì¸ ê²½ìš°
                else:
                    print(f"ğŸ‰ ì •ë‹µ! ğŸ‰")  # ì •ë‹µ ë©”ì‹œì§€ ì¶œë ¥
                    guessed_correctly = True  # í”Œë˜ê·¸ë¥¼ Trueë¡œ ì„¤ì •í•˜ì—¬ ë°˜ë³µ ì¢…ë£Œ
            except ValueError:  # ìˆ«ìê°€ ì•„ë‹Œ ê°’ì„ ì…ë ¥í–ˆì„ ë•Œ ë°œìƒí•˜ëŠ” ì˜ˆì™¸ ì²˜ë¦¬
                print("âš ï¸ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”. ë¬¸ìëŠ” ì…ë ¥í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")  # ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥

        # ì •ë‹µì„ ë§ì¶˜ í›„, ê²Œì„ ì¬ì‹œì‘ ì—¬ë¶€ë¥¼ ë¬¼ìŒ
        play_again = input("í•œ ê²Œì„ ë”? (y/n): ").strip().lower()  # ê³µë°± ì œê±° ë° ì†Œë¬¸ì ë³€í™˜
        if play_again != 'y':  # 'y'ê°€ ì•„ë‹ˆë©´ ê²Œì„ ì¢…ë£Œ
            print("ì¢…ë£Œ!")  # ì¢…ë£Œ ë©”ì‹œì§€ ì¶œë ¥
            break  # ë°”ê¹¥ ë£¨í”„ ì¢…ë£Œ

# ê²Œì„ ì‹¤í–‰
number_game()  # number_game í•¨ìˆ˜ í˜¸ì¶œ
```

```python
#1. ìˆ«ì ë§ì¶”ê¸° ê²Œì„
import random  # ëœë¤ ìˆ«ìë¥¼ ìƒì„±í•˜ê¸° ìœ„í•´ random ëª¨ë“ˆ ê°€ì ¸ì˜¤ê¸°

# 1ë¶€í„° 10 ì‚¬ì´ì˜ ëœë¤ ìˆ«ì ìƒì„±
ans = random.randint(1, 10)  # ì •ë‹µ ìˆ«ì ì„¤ì •
print("1ë¶€í„° 10 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”!")  # ê²Œì„ ì‹œì‘ ë©”ì‹œì§€ ì¶œë ¥

# ì •ë‹µì„ ë§ì¶œ ë•Œê¹Œì§€ ë°˜ë³µ
while True:
    try:
        # ì‚¬ìš©ìë¡œë¶€í„° ìˆ«ìë¥¼ ì…ë ¥ë°›ê¸°
        num = int(input("ì´ ìˆ«ìëŠ” ë¬´ì—‡ì¼ê¹Œìš”? : "))  # ì…ë ¥ê°’ì„ ì •ìˆ˜ë¡œ ë³€í™˜
        
        # ì…ë ¥ê°’ì´ ì •ë‹µë³´ë‹¤ ì‘ì€ ê²½ìš°
        if num < ans:
            print(f"{num}ì€(ëŠ”) ì‘ë‹¤!")  # ì‘ë‹¤ëŠ” íŒíŠ¸ ì œê³µ
        
        # ì…ë ¥ê°’ì´ ì •ë‹µë³´ë‹¤ í° ê²½ìš°
        elif num > ans:
            print(f"{num}ì€(ëŠ”) í¬ë‹¤!")  # í¬ë‹¤ëŠ” íŒíŠ¸ ì œê³µ
        
        # ì…ë ¥ê°’ì´ ì •ë‹µê³¼ ê°™ì€ ê²½ìš°
        else:
            print("ì •ë‹µ!")  # ì •ë‹µ ë©”ì‹œì§€ ì¶œë ¥
            break  # ë£¨í”„ ì¢…ë£Œ (ê²Œì„ ì¢…ë£Œ)
    
    except ValueError:  # ìˆ«ìê°€ ì•„ë‹Œ ê°’ì´ ì…ë ¥ëœ ê²½ìš° ì˜ˆì™¸ ì²˜ë¦¬
        print("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”!")  # ì‚¬ìš©ìì—ê²Œ ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥
```

```python
# 2. í´ë˜ìŠ¤ì™€ í•¨ìˆ˜ ì‚¬ìš©í•˜ê¸°
import re  # ì •ê·œ í‘œí˜„ì‹ì„ ì‚¬ìš©í•˜ê¸° ìœ„í•œ re ëª¨ë“ˆ ê°€ì ¸ì˜¤ê¸°

# ì´ë¦„ ì…ë ¥ í•¨ìˆ˜ ì •ì˜
def get_name():
    while True:  # ì˜¬ë°”ë¥¸ ì´ë¦„ì´ ì…ë ¥ë  ë•Œê¹Œì§€ ë°˜ë³µ
        name = input("ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ").strip()  # ì´ë¦„ ì…ë ¥ë°›ê³  ì•ë’¤ ê³µë°± ì œê±°
        if re.match(r"^[ê°€-í£\s]+$", name):  # í•œê¸€ ì´ë¦„(ê°€-í£ê³¼ ê³µë°±)ì´ ë§ëŠ”ì§€ ê²€ì‚¬
            return name  # ì˜¬ë°”ë¥¸ í•œê¸€ ì´ë¦„ì´ë©´ ë°˜í™˜
        elif re.match(r"^[a-zA-Z\s]+$", name):  # ì˜ì–´ ì´ë¦„(ì•ŒíŒŒë²³ê³¼ ê³µë°±)ì´ ë§ëŠ”ì§€ ê²€ì‚¬
            return name.title()  # ì˜ì–´ ì´ë¦„ì´ë©´ ì²« ê¸€ìë¥¼ ëŒ€ë¬¸ìë¡œ ë³€í™˜í•˜ì—¬ ë°˜í™˜
        else:  # ë‘˜ ë‹¤ ì•„ë‹Œ ê²½ìš°
            print("ì´ë¦„ì„ ì…ë ¥í•˜ë¼ë‹ˆê¹Œ.")  # ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥

# ì„±ë³„ ì…ë ¥ í•¨ìˆ˜ ì •ì˜
def get_gender():
    while True:  # ì˜¬ë°”ë¥¸ ì„±ë³„ì´ ì…ë ¥ë  ë•Œê¹Œì§€ ë°˜ë³µ
        gender = input("ì„±ë³„ì„ ì…ë ¥í•˜ì„¸ìš” (m/f ë˜ëŠ” male/female):").lower()  # ì†Œë¬¸ìë¡œ ë³€í™˜
        if gender in ['f', 'female']:  # ì—¬ì„±(Female) ì…ë ¥ í™•ì¸
            return 'Female'  # í‘œì¤€í™”ëœ í˜•ì‹ìœ¼ë¡œ ë°˜í™˜
        elif gender in ['m', 'male']:  # ë‚¨ì„±(Male) ì…ë ¥ í™•ì¸
            return 'Male'  # í‘œì¤€í™”ëœ í˜•ì‹ìœ¼ë¡œ ë°˜í™˜
        else:  # ì˜ëª»ëœ ì…ë ¥ì¸ ê²½ìš°
            print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤.\nm/f ë˜ëŠ” male/female ì¤‘ì—ì„œ ì„ íƒí•˜ì„¸ìš”.")  # ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥

# ë‚˜ì´ ì…ë ¥ í•¨ìˆ˜ ì •ì˜
def get_age():
    while True:  # ì˜¬ë°”ë¥¸ ë‚˜ì´ê°€ ì…ë ¥ë  ë•Œê¹Œì§€ ë°˜ë³µ
        try:
            age = int(input("ë‚˜ì´ë¥¼ ì…ë ¥í•˜ì„¸ìš”:"))  # ë‚˜ì´ë¥¼ ì •ìˆ˜ë¡œ ì…ë ¥ë°›ìŒ
            if 0 <= age <= 120:  # ë‚˜ì´ê°€ 0~120 ë²”ìœ„ ë‚´ì¸ì§€ í™•ì¸
                return age  # ì˜¬ë°”ë¥¸ ë‚˜ì´ë©´ ë°˜í™˜
            elif age > 120:  # ì…ë ¥ê°’ì´ 120ë³´ë‹¤ í° ê²½ìš°
                print("ê±°ì§“ë§!!!")  # ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥
            else:  # ì…ë ¥ê°’ì´ 0ë³´ë‹¤ ì‘ì€ ê²½ìš°
                print("ë‚˜ì´ë¥¼ ê±°ê¾¸ë¡œ ë“œì…¨êµ°ìš”.. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.")  # ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥
        except ValueError:  # ìˆ«ìê°€ ì•„ë‹Œ ê°’ì„ ì…ë ¥í•œ ê²½ìš° ì˜ˆì™¸ ì²˜ë¦¬
            print("ë‚˜ì´ëŠ” ìˆ«ìì…ë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•˜ì„¸ìš”.")  # ê²½ê³  ë©”ì‹œì§€ ì¶œë ¥

# Person í´ë˜ìŠ¤ ì •ì˜
class Person:
    def __init__(self, name, gender, age):  # ê°ì²´ ìƒì„± ì‹œ í˜¸ì¶œë˜ëŠ” ìƒì„±ì
        self.name = name  # ì´ë¦„ ì €ì¥
        self.gender = gender  # ì„±ë³„ ì €ì¥
        self.age = age  # ë‚˜ì´ ì €ì¥
    
    def display(self):  # ê°ì²´ ì •ë³´ë¥¼ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ
        print(f"Name: {self.name}, Gender: {self.gender}")  # ì´ë¦„ê³¼ ì„±ë³„ ì¶œë ¥
        print(f"Age: {self.age}")  # ë‚˜ì´ ì¶œë ¥

# ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
name = get_name()  # ì´ë¦„ ì…ë ¥ë°›ê¸° í•¨ìˆ˜ í˜¸ì¶œ
gender = get_gender()  # ì„±ë³„ ì…ë ¥ë°›ê¸° í•¨ìˆ˜ í˜¸ì¶œ
age = get_age()  # ë‚˜ì´ ì…ë ¥ë°›ê¸° í•¨ìˆ˜ í˜¸ì¶œ

# Person ê°ì²´ ìƒì„± ë° ì´ˆê¸°í™”
person = Person(name, gender, age)  # ì…ë ¥ë°›ì€ ì •ë³´ë¥¼ ê¸°ë°˜ìœ¼ë¡œ Person ê°ì²´ ìƒì„±

# Person ê°ì²´ ì •ë³´ ì¶œë ¥
person.display()  # display ë©”ì„œë“œë¥¼ í˜¸ì¶œí•˜ì—¬ ê°ì²´ ì •ë³´ ì¶œë ¥
```
