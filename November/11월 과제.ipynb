```python
# Person 클래스 정의
class Person:
    def __init__(self, name, gender, age):  # 생성자: Person 객체를 초기화하는 메서드
        self.name = name  # 입력받은 이름을 객체의 속성으로 저장
        self.gender = gender.capitalize()  # 성별을 첫 글자만 대문자로 변환하여 저장 ('male' -> 'Male')
        self.age = age  # 입력받은 나이를 객체의 속성으로 저장

    def display(self):  # 객체 정보를 출력하는 메서드
        print(f"이름: {self.name}, 성별: {self.gender}")  # 이름과 성별을 출력
        print(f"나이: {self.age}")  # 나이를 출력

    def greet(self):  # 나이에 따라 다른 인사말을 출력하는 메서드
        if self.age < 13:  # 나이가 13세 미만인 경우
            print(f"안녕하세요, {self.name}어린이! 반가워요!")  # 어린이에 대한 인사말 출력
        elif 13 <= self.age <= 19:  # 나이가 13~19세인 경우
            print(f"안녕하세요, {self.name}님! 청소년이시군요!")  # 청소년에 대한 인사말 출력
        elif 20 <= self.age <= 59:  # 나이가 20~59세인 경우
            print(f"안녕하세요, {self.name}님! 성인이시군요!")  # 성인에 대한 인사말 출력
        else:  # 나이가 60세 이상인 경우
            print(f"안녕하세요, {self.name}님! 어르신이시군요!")  # 어르신에 대한 인사말 출력


# 사용자 입력 받기
name = get_name()  # 이름 입력받기 함수 호출 (유효성 검사는 get_name 함수에서 처리)
gender = get_gender()  # 성별 입력받기 함수 호출 (유효성 검사는 get_gender 함수에서 처리)
age = get_age()  # 나이 입력받기 함수 호출 (유효성 검사는 get_age 함수에서 처리)

# Person 객체 생성 및 동작
person = Person(name, gender, age)  # 입력받은 데이터를 기반으로 Person 객체 생성
person.display()  # 객체 정보 출력 메서드 호출
person.greet()  # 나이에 따라 다른 인사말을 출력하는 메서드 호출
```

```python
import random  # 랜덤 숫자 생성을 위한 random 모듈 가져오기

# 숫자 맞추기 게임 함수 정의
def number_game():
    # 게임 시작 안내 메시지 출력
    print("숫자 맞추기 게임에 오신 것을 환영합니다!\n1부터 10 사이의 숫자를 맞춰보세요.")

    while True:  # 사용자가 게임을 끝낼 때까지 반복
        # 1부터 10 사이의 랜덤 숫자 생성
        target_number = random.randint(1, 10)  # 정답 숫자 생성
        guessed_correctly = False  # 정답을 맞췄는지 여부를 나타내는 플래그 변수

        while not guessed_correctly:  # 정답을 맞출 때까지 반복
            try:
                # 사용자로부터 숫자를 입력받음
                user_input = int(input("숫자를 입력하세요: "))  # 입력값을 정수로 변환
                # 입력값이 1~10 범위를 벗어난 경우
                if user_input < 1 or user_input > 10:
                    print("⚠️ 1부터 10 사이의 숫자를 입력하세요.")  # 범위 경고 메시지 출력
                # 입력값이 정답보다 작은 경우
                elif user_input < target_number:
                    print("이것보단 커요!")  # 힌트 제공
                # 입력값이 정답보다 큰 경우
                elif user_input > target_number:
                    print("이것보단 작아요!")  # 힌트 제공
                # 입력값이 정답인 경우
                else:
                    print(f"🎉 정답! 🎉")  # 정답 메시지 출력
                    guessed_correctly = True  # 플래그를 True로 설정하여 반복 종료
            except ValueError:  # 숫자가 아닌 값을 입력했을 때 발생하는 예외 처리
                print("⚠️ 숫자를 입력하세요. 문자는 입력할 수 없습니다.")  # 경고 메시지 출력

        # 정답을 맞춘 후, 게임 재시작 여부를 물음
        play_again = input("한 게임 더? (y/n): ").strip().lower()  # 공백 제거 및 소문자 변환
        if play_again != 'y':  # 'y'가 아니면 게임 종료
            print("종료!")  # 종료 메시지 출력
            break  # 바깥 루프 종료

# 게임 실행
number_game()  # number_game 함수 호출
```

```python
#1. 숫자 맞추기 게임
import random  # 랜덤 숫자를 생성하기 위해 random 모듈 가져오기

# 1부터 10 사이의 랜덤 숫자 생성
ans = random.randint(1, 10)  # 정답 숫자 설정
print("1부터 10 사이의 숫자를 맞춰보세요!")  # 게임 시작 메시지 출력

# 정답을 맞출 때까지 반복
while True:
    try:
        # 사용자로부터 숫자를 입력받기
        num = int(input("이 숫자는 무엇일까요? : "))  # 입력값을 정수로 변환
        
        # 입력값이 정답보다 작은 경우
        if num < ans:
            print(f"{num}은(는) 작다!")  # 작다는 힌트 제공
        
        # 입력값이 정답보다 큰 경우
        elif num > ans:
            print(f"{num}은(는) 크다!")  # 크다는 힌트 제공
        
        # 입력값이 정답과 같은 경우
        else:
            print("정답!")  # 정답 메시지 출력
            break  # 루프 종료 (게임 종료)
    
    except ValueError:  # 숫자가 아닌 값이 입력된 경우 예외 처리
        print("숫자를 입력하세요!")  # 사용자에게 경고 메시지 출력
```

```python
# 2. 클래스와 함수 사용하기
import re  # 정규 표현식을 사용하기 위한 re 모듈 가져오기

# 이름 입력 함수 정의
def get_name():
    while True:  # 올바른 이름이 입력될 때까지 반복
        name = input("이름을 입력하세요: ").strip()  # 이름 입력받고 앞뒤 공백 제거
        if re.match(r"^[가-힣\s]+$", name):  # 한글 이름(가-힣과 공백)이 맞는지 검사
            return name  # 올바른 한글 이름이면 반환
        elif re.match(r"^[a-zA-Z\s]+$", name):  # 영어 이름(알파벳과 공백)이 맞는지 검사
            return name.title()  # 영어 이름이면 첫 글자를 대문자로 변환하여 반환
        else:  # 둘 다 아닌 경우
            print("이름을 입력하라니까.")  # 경고 메시지 출력

# 성별 입력 함수 정의
def get_gender():
    while True:  # 올바른 성별이 입력될 때까지 반복
        gender = input("성별을 입력하세요 (m/f 또는 male/female):").lower()  # 소문자로 변환
        if gender in ['f', 'female']:  # 여성(Female) 입력 확인
            return 'Female'  # 표준화된 형식으로 반환
        elif gender in ['m', 'male']:  # 남성(Male) 입력 확인
            return 'Male'  # 표준화된 형식으로 반환
        else:  # 잘못된 입력인 경우
            print("잘못된 입력입니다.\nm/f 또는 male/female 중에서 선택하세요.")  # 경고 메시지 출력

# 나이 입력 함수 정의
def get_age():
    while True:  # 올바른 나이가 입력될 때까지 반복
        try:
            age = int(input("나이를 입력하세요:"))  # 나이를 정수로 입력받음
            if 0 <= age <= 120:  # 나이가 0~120 범위 내인지 확인
                return age  # 올바른 나이면 반환
            elif age > 120:  # 입력값이 120보다 큰 경우
                print("거짓말!!!")  # 경고 메시지 출력
            else:  # 입력값이 0보다 작은 경우
                print("나이를 거꾸로 드셨군요.. 다시 입력해주세요.")  # 경고 메시지 출력
        except ValueError:  # 숫자가 아닌 값을 입력한 경우 예외 처리
            print("나이는 숫자입니다. 다시 입력하세요.")  # 경고 메시지 출력

# Person 클래스 정의
class Person:
    def __init__(self, name, gender, age):  # 객체 생성 시 호출되는 생성자
        self.name = name  # 이름 저장
        self.gender = gender  # 성별 저장
        self.age = age  # 나이 저장
    
    def display(self):  # 객체 정보를 출력하는 메서드
        print(f"Name: {self.name}, Gender: {self.gender}")  # 이름과 성별 출력
        print(f"Age: {self.age}")  # 나이 출력

# 사용자 입력 받기
name = get_name()  # 이름 입력받기 함수 호출
gender = get_gender()  # 성별 입력받기 함수 호출
age = get_age()  # 나이 입력받기 함수 호출

# Person 객체 생성 및 초기화
person = Person(name, gender, age)  # 입력받은 정보를 기반으로 Person 객체 생성

# Person 객체 정보 출력
person.display()  # display 메서드를 호출하여 객체 정보 출력
```
